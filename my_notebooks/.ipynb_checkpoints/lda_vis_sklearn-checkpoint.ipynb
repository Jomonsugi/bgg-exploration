```python
import pyLDAvis
import pyLDAvis.sklearn
pyLDAvis.enable_notebook()
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pickle
```

```python
with open('../data/doc_lst.pkl', 'rb') as fp:
    processed = pickle.load(fp)
print(type(processed))
print(processed[0])
```

```python
tf_vectorizer = CountVectorizer(
                                max_df = 0.2, 
                                min_df = 20)
desc_tf = tf_vectorizer.fit_transform(processed)
print(desc_tf.shape)
print(type(desc_tf))
```

```python
tfidf_vectorizer = TfidfVectorizer(**tf_vectorizer.get_params())
desc_tfidf = tfidf_vectorizer.fit_transform(processed)
print(desc_tfidf.shape)
```

```python
# for TF DTM
lda_tf = LatentDirichletAllocation(n_topics=100, random_state=34)
lda_tf.fit(desc_tfidf)
# for TFIDF DTM
lda_tfidf = LatentDirichletAllocation(n_topics=100, random_state=34)
lda_tfidf.fit(desc_tfidf)
```

```python
pyLDAvis.sklearn.prepare(lda_tf, desc_tfidf, tf_vectorizer)

```

```python
with open('../data/comments_doc_lst.pkl', 'rb') as fp:
    c_processed = pickle.load(fp)
print(type(c_processed))
```

```python
tf_vectorizer = CountVectorizer(
                                max_df = 0.4, 
                                min_df = 10)
comments_tf = tf_vectorizer.fit_transform(c_processed)
print(comments_tf.shape)
print(type(comments_tf))
```

```python
# for TF DTM
lda_tf = LatentDirichletAllocation(n_topics=25, random_state=34)
lda_tf.fit(comments_tfidf)
# for TFIDF DTM
lda_tfidf = LatentDirichletAllocation(n_topics=25, random_state=34)
lda_tfidf.fit(comments_tfidf)
```

```python
pyLDAvis.sklearn.prepare(lda_tf, comments_tfidf, tf_vectorizer)
```
